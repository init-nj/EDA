### Methodology

The analysis follows the **UBM framework**:
- **Univariate Analysis** — Examining one variable at a time (e.g., order status distribution, category counts).
- **Bivariate Analysis** — Exploring relationships between two variables (e.g., gender vs. amount, state vs. revenue).
- **Multivariate Analysis** — Understanding interactions among multiple variables simultaneously (e.g., category vs. quantity vs. amount, geographic hierarchy).

## 3. ***Data Wrangling***

### Data Wrangling Code

````python
# Write your code to make your dataset analysis ready.
# Step 1: Make a backup copy of the original data
df = sales_data.copy()
# Step 2: Drop the 'index' column — it's just a row number, adds no analytical value
df.drop(columns=['index'], inplace=True, errors='ignore')
# B2B should be True/False (boolean), not "True"/"False" text
df['B2B'] = df['B2B'].astype('bool')
df['Date'] = pd.to_datetime(df['Date'], errors='coerce')
categorical_cols = [
    'Gender', 'Age Group', 'Status', 'Channel',
    'Category', 'Size', 'ship-city',
    'ship-state', 'ship-country', 'B2B'
]
for col in categorical_cols:
    df[col] = df[col].astype('category')
# Step 4: Standardize text fields
# "Maharashtra" and "maharashtra" are the same state — lowercase + strip spaces fixes this
text_cols = ['ship-city', 'ship-state']
for col in text_cols:
    df[col] = df[col].str.lower().str.strip()
df['Revenue_per_item'] = df['Amount'] / df['Qty']

# Extract time-based features from the Date column 
df['Day']        = df['Date'].dt.day          
df['Weekday']    = df['Date'].dt.day_name()   
df['Month_name'] = df['Date'].dt.month_name() 

print("Shape after wrangling:", df.shape)
print(df.info())
````

### What all manipulations have you done and insights you found?
### Manipulations Performed

Applied several data wrangling steps to make the dataset analysis-ready:

* **Removed unnecessary column**: Dropped the `index` column as it had no analytical value.
* **Data type corrections**: Converted `Date` to datetime format and categorical columns (like Gender, Category, Channel, etc.) to appropriate types for efficient analysis.
* **Standardization**: Cleaned text fields such as `ship-city` and `ship-state` by converting them to lowercase and removing extra spaces to avoid duplicate categories.
* **Outlier handling**: Filtered unrealistic age values (restricted between 18–70) to improve data reliability.
* **Feature engineering**: Created new columns like `Revenue_per_item`, `Day`, `Weekday`, and `Month_name` to enable deeper time-based and pricing analysis.
* **Data type refinement**: Converted `B2B` column into boolean for clarity and consistency.

---

### Insights Found

* Most orders have **quantity = 1**, indicating single-item purchasing behavior.
* The **average transaction value (~₹682)** suggests mid-range products dominate sales.
* Customers are primarily in the **working-age group (28–47 years)**.
* Sales data spans multiple locations, showing **wide geographic distribution**.
* Feature engineering enables identifying trends like **peak sales days, top-performing categories, and regional demand patterns**.

````python
# Chart - 7 visualization code
# A scatter plot showing the relationship between Quantity ordered and Total Amount paid
# Each dot is one order, colored and shaped by its delivery Status
# This is a Bivariate (Numerical vs Numerical) chart with a categorical overlay
sns.set_theme(style='whitegrid')
plt.figure(figsize=(10, 6))
sns.scatterplot(
    data=df,
    x='Qty',        
    y='Amount',     
    hue='Status',   
    style='Status', 
    alpha=0.6,      
    palette='deep'
)
plt.title('Relationship between Quantity and Order Amount')
plt.xlabel('Quantity Ordered')
plt.ylabel('Total Amount')
plt.show()
````

##### 1. Why did you pick the specific chart?

A scatter plot was chosen because we are comparing **two continuous numerical variables** (Quantity and Amount) against each other. Scatter plots are the standard tool for spotting correlations, clusters, and outliers between two numeric columns. Adding `hue='Status'` as a color layer lets us simultaneously see whether order status (Delivered, Cancelled, Returned) follows any particular pattern in the Qty–Amount space — something a simple line or bar chart cannot show.


##### 2. What is/are the insight(s) found from the chart?

- **Orders cluster at low quantity (1–2 items)** with amounts mostly between ₹300–₹1500, confirming the single-item purchase behavior seen earlier.

## **5. Solution to Business Objective**

#### What do you suggest the client to achieve Business Objective ?
Explain Briefly.

Based on the EDA, here are the key recommendations for Vrinda Store to achieve its business objectives:

**1. Double Down on Top Channels**
Amazon, Myntra, and Flipkart are driving the majority of orders. The store should prioritize listings, sponsored ads, and promotions on these platforms. Low-performing channels (Nalli, Others) should either be optimized or deprioritized to reduce operational overhead.

**2. Boost the Underperforming Product Category**
"Bottoms" consistently sell the least. The store should investigate whether this is a pricing issue, a visibility issue, or a product-market fit problem. Bundling Bottoms with top-performing Sets at a slight discount could improve their sales while boosting overall cart value.

**3. Localize Marketing for Non-Hindi Speaking States**
States like Tamil Nadu, Kerala, and West Bengal show lower order volumes. This signals a lack of language-inclusive marketing. The store should run ads and product descriptions in regional languages to tap into this underserved market.

**4. Reduce Returns and Cancellations**
The combined 6.1% return and cancellation rate is a revenue drain. The store should invest in better product photography, accurate sizing guides, and faster shipping SLAs. A post-delivery feedback loop would help identify the root cause (wrong size, damaged goods, late delivery) and fix it systematically.

**5. Maximize Peak Day Sales**
With Friday being the strongest sales day, the store should schedule flash sales, email blasts, and push notifications for Friday mornings. For Monday (weakest day), exclusive "Monday Deals" with loyalty points can help drive traffic on slow days.

**6. Target the Core Customer Demographic**
The 28–47 age group, with a slight male spending edge, is the core audience. Marketing content (visuals, copy, influencer choices) should reflect this demographic's preferences and lifestyle to maximize conversion rates.

**7. Strengthen Geographic Strongholds**
Maharashtra, Karnataka, and Uttar Pradesh are the top-grossing states. The store should ensure fast delivery timelines and local warehousing in these regions to maintain customer satisfaction and repeat purchase rates.
